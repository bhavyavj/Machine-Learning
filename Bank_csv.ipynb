```python
m=0
f=0
#for Male in df['Gender']:
 #       m=m+1
for i in df['Gender']:
    if i=='Male':
        m=m+1
    else:
        f=f+1
    
#print(m,f)

Gender=['Male','Female']
NMF=[m,f]
plt.pie(NMF,labels=Gender, shadow=True,autopct='%1.1f')
```

```python
fr=0
sp=0
ger=0
#for Male in df['Gender']:
 #       m=m+1
for i in df['Geography']:
    if i=='France':
        fr+=1
    elif i=='Spain':
        sp+=1
    else:
        ger+=1
    
#print(m,f)

Gender=['France','Spain','Germany']
NMF=[fr,sp,ger]
plt.pie(NMF,labels=Gender, shadow=True,autopct='%1.1f')
```

```python
hcc=0
dhcc=0
#for Male in df['Gender']:
 #       m=m+1
for i in df['HasCrCard']:
    if i==1:
        hcc+=1
    else:
        dhcc+=1
    
#print(m,f)

Gender=['HasCrCard','NoCrCard']
NMF=[hcc,dhcc]
plt.pie(NMF,labels=Gender, shadow=True,autopct='%1.1f')
```
