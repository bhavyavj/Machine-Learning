```
#reading from a file 
f=open('mywebdata.txt','r')
mydata=f.read()
f.close()
```

```
newdata1=[i for i in mydata.split()]
```

```
# NLTK graphs 
import matplotlib.pyplot as plt
import nltk
%matplotlib inline # incase matplotlib is not  supporting by default with pandas and nltk
# to replace plt.show() from any platform
```

```
[i for i in dir(nltk) if 'Freq' in i] # this will plot data for
```

```
nlpdata=nltk.FreqDist(newdata1)
```

```
nlpdata.plot(20) # top 20 
```

```
#we can remove stop words from newdata1 the plot the graph
from nltk.corpus import  stopwords
nltk.download('stopwords')
#removing stopwords
```

```
removedata=[i for i in newdata1 if i.lower() not in stopwords.words('english')]
nlpremove=nltk.FreqDist(removedata)
nlpremove.plot(15)
```
