```
import matplotlib.pyplot as plt  # only loading python lib
```

```
x=[2,3]
y=[9,5]

plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y)
plt.show()
```

```
x=[2,3]
y=[9,5]

plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y)
plt.grid(color="green")  # to form grid in graph
plt.show()
```

```
x=[2,3]
y=[9,5]
x1=[4,3,8]
y1=[2,9,7]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y,x1,y1)
plt.grid(color="green")  # to form grid in graph
plt.show()
```

```
x=[2,3]
y=[9,5]
x1=[4,3,8]
y1=[2,9,7]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y,label="water")
plt.plot(x1,y1,label="sand")
plt.grid(color="green")  # to form grid in graph
plt.legend()
plt.show()
```

```
x=[2,3]
y=[9,5]
x1=[4,3,8]
y1=[2,9,7]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y,label="water")  #this will draw a straight line
plt.plot(x1,y1,label="sand")
plt.grid(color="green")  # to form grid in graph
plt.legend()   #to show label plot
plt.xlim(0,12)  # to show min and max of x
plt.ylim(0,15)  # to show y axis limit
plt.show()      
```

```
x=[2,3]
y=[9,5]
x1=[4,3,8]
y1=[2,9,7]
plt.xlabel("time")
plt.ylabel("speed")
plt.bar(x,y)
plt.bar(x1,y1)
plt.plot(x,y,label="water")
plt.plot(x1,y1,label="sand")
plt.grid(color="green")  # to form grid in graph
plt.legend()
plt.show()
```
